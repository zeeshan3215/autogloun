```python
import pandas as pd
import numpy as np
df = pd.read_csv("Cars_India_dataset.csv")
df.head(5)
```

```python
from pandas.core.reshape.reshape import get_dummies

from sklearn.preprocessing import LabelEncoder
y1=LabelEncoder().fit_transform(df['Model'])
y=pd.get_dummies(y1).values
```

```python
y
```

```python
data=df[['Displacement','Length','Width','Height','Wheelbase','Model']]

```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(data,test_size=0.2)
```

```python
test1=test.drop(['Model'],axis=1)
train
```

```python
from autogluon.tabular import TabularPredictor
predictor= TabularPredictor( label ='Model').fit(train_data = train, verbosity = 2,presets='best_quality')
```

```python
test
```

```python
predictor.leaderboard(train, silent=True)
```

```python
predictions = predictor.predict(test1)
predictions
```

```python
test_score = predictor.evaluate(test)
print(test_score)
```
